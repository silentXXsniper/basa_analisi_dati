<h2 style="font-family: 'Georgia'; color: #34495e; border-bottom: 3px solid #3498db; padding-bottom: 10px;">Introduzione a H5PY e il formato .h5</h2>
<p style="font-family: 'Arial'; font-size: 16px; color: #2c3e50; margin-top: 20px;"> H5py consente di accedere ai file messi a disposizione della nasa, un formato comune nel campo scientifico</p>

```python
import h5py
import numpy as np
```

<h2 style="font-family: 'Georgia'; color: #34495e; border-bottom: 3px solid #3498db; padding-bottom: 10px;">come esplorare i file</h2>
<p style="font-family: 'Arial'; font-size: 16px; color: #2c3e50; margin-top: 20px;"> per accedervi Ã¨ necessario l'installazione del pip h5py, con esso Ã¨ possibile accedere alla struttura interna del file, il quale Ã¨ strutturato con una serie di sottocartelle chiamate keys.</p>    
<p style="font-family: 'Arial'; font-size: 16px; color: #2c3e50; margin-top: 20px;">Un file .h5 puÃ² avere un solo file all'interno (quindi nessuna sottocartella) o essere formato da varie ramificazioni anche molto complesse.</p>

```python
path='datasets/DATASET_NASA_2020/SMAP_L3_SM_P_E_20200101_R19240_001.h5'  #scegliamo arbitrariamente un dataset da aprire
with h5py.File(path,'r') as hdf:                                         #il file viene aperto con l'apposito comando, con 'r' si intende read
    ls=list(hdf.keys())                                                  #il comando crea una variabile contentenente tutte le keys di primo livello
    print('List of datests in this file: \n', ls)                        #le stampa per poter essere lette dall'utente
    data= hdf.get('/Soil_Moisture_Retrieval_Data_AM/soil_moisture')      #crea un array dei dati presenti (a volte non sufficenti da soli)
    dataset1 = np.array(data)                                            #crea un dataset
    print('Shape of dataset1: \n', dataset1.shape)                       #ne analizza la forma (a scopo dimostrativo)
```
